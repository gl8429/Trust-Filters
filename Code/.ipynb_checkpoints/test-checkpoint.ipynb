```python
import os
import json
import numpy as np
import nltk
import langdetect

from pprint import pprint
from bs4 import BeautifulSoup
from nltk.corpus import stopwords

PATH = '/Users/Lucifer/Documents/GraduateStudy/NLP/Trust-Filters/Code/new_schema/71851391.json'

with open(PATH) as data_file:    
    data = json.load(data_file)

corpus = []
```

```python
pprint(data['posts'][0]['text'])
```

```python
for i in range(len(data['posts'])):
    post = data['posts'][i]['text']
    content = []
    for sentences in post.split('\n'):
        soup = BeautifulSoup(sentences, 'lxml')
        for el in soup.find_all('p', text = True):
            if el.string[0] == '©' or el.string[:14] == "Photo credit: ":
                break
            content.append(el.string)
    tokens = " ".join(content)
    corpus.append(tokens.split(' '))
```

```python
class CleanTokens(object):

    def __init__(self):
        self._stopwords = stopwords.words('english')

    def lowercase(self, tokens):
        tokens = [token.lower() for token in tokens]
        return tokens

    def punctuation(self, tokens):
        tokens = [token for token in tokens if token.isalpha()]
        return tokens

    def stopwords(self, tokens):
        tokens = [token for token in tokens if token not in self._stopwords]
        return tokens

    def english(self, tokens):
        if tokens:
            lang = langdetect.detect(' '.join(tokens))
            if lang == 'en':
                return tokens
        else:
            return []

    def normalize(self, tokens):
#         tokens = self.english(tokens)
        if tokens:
            tokens = self.punctuation(tokens)
            tokens = self.lowercase(tokens)
            tokens = self.stopwords(tokens)
            return tokens
        else:
            return []
```

```python
x = CleanTokens()
cleanToken = []
for token in corpus:
    cleanToken.append(x.normalize(token))
```

```python
print(cleanToken)
```

```python
from nltk import stem
porter = stem.porter.PorterStemmer()
cleanStem = []
for token in cleanToken:
    cleanStem.append([porter.stem(i) for i in token])
```
